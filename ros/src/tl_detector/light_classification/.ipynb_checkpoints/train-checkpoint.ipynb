```python
import csv
import cv2
import numpy as np
from keras.callbacks import ModelCheckpoint, Callback
import random
import sklearn
from sklearn.model_selection import train_test_split
from keras.models import Sequential, Model
from keras.layers import Dense, Flatten,Cropping2D
from keras.layers import Conv2D, GlobalAveragePooling2D
from keras.layers import Lambda
from keras.regularizers import l2
# from keras.applications import MobileNet
from keras.applications import VGG16
```

```python
lines =[]
# Load the data
with open('/home/paul/Documents/Udacity/CarND-Capstone/imgs/Simulator_Dataset/data_list.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        lines.append(line)
    print('Udacity simulator data loaded.')
    
```

```python
train_samples, validation_and_test_samples = train_test_split(lines, test_size=0.2)
validation_samples, test_samples = train_test_split(validation_and_test_samples, test_size=0.5)
print'Number of whole samples: ', len(lines)
print'Number of train_samples: ', len(train_samples)
print'Number of validation_samples: ', len(validation_samples)
print'Number of test_samples: ', len(test_samples)
```

```python
base_model=VGG16(input_shape=(600, 800, 3), weights='imagenet', include_top=False) #imports the VGG16 model and discards the last 1000 neuron layer.

x=base_model.output
x=GlobalAveragePooling2D()(x)
x=Dense(512,activation='relu')(x) #dense layer 3
preds=Dense(1)(x) #final layer with softmax activation
```

```python
model=Model(inputs=base_model.input,outputs=preds)
print(model.summary())
for i,layer in enumerate(model.layers):
    print(i,layer.name)
```

```python
# adapt thee VGG16 model

#we want to set the first 18 layers of the network to be non-trainable
for layer in model.layers[:18]:
    layer.trainable=False
for layer in model.layers[19:]:
    layer.trainable=True
```

```python
#define the help functions
def createPreProcessingLayers():
    '''
     This funciton is used to bild a initial model, where the images would preprocessed.
     '''
    model = Sequential()
    model.add(Lambda(lambda x: (x / 255.0) - 0.5, input_shape=(600,800,3)))
    model.add(Cropping2D(cropping=((50,20), (0,0))))
    return model
    
def random_darken(img):
    '''
     Function to distort the dataset through adding random shadow to the iamge
     '''
    #Convert original images from RGB colorspace to HSV colorspace
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    #Generate new random brightness
    rand = random.uniform(0.5,1.0)
    hsv[:,:,2] = rand*hsv[:,:,2]
    #Convert images back to RGB colorspace
    new_img = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
    return new_img 
    
def generator(samples, batch_size=32,validation=False):
    '''
    This function gereate the data for training and validation.
    The validationsdata will darken the images randomly and add change some random noise on the angle.
    '''
    num_samples = len(samples)
    while True: # Loop forever so the generator never terminates
        random.shuffle(samples)
        for offset in range(0, num_samples, batch_size):
            batch_samples = samples[offset:offset+batch_size]
            images = []
            colors = []
            for batch_sample in batch_samples:
                name_image = batch_sample[0].strip()
                color = int(batch_sample[1].strip())
                if validation:
                    image = cv2.imread(name_image)                             
                else:
                    image = random_darken(cv2.imread(name_image)) 
                images.append(image)
                colors.append(color)
            # trim image to only see section with road
            x_train = np.array(images)
            y_train = np.array(colors)
            '''
            print "shape x_train: ", x_train.shape
            # print x_train.shape()[1]
            print "shape y_train: ", y_train.shape
            # print y_train.shape()[1]
            '''
            yield sklearn.utils.shuffle(x_train, y_train)
```

```python
# compile and train the model using the generator function
model.compile(optimizer = 'adam', loss = 'mse', metrics=['accuracy'])    
train_generator = generator(train_samples, batch_size=16, validation=False)

validation_generator = generator(validation_samples, batch_size=16, validation=True)

test_generator = generator(test_samples, batch_size=16, validation=True)

#define the checkpoint for each epoch
checkpoint = ModelCheckpoint('model{epoch:02d}.h5',monitor='val_loss', verbose=1,\
                             save_best_only=False, save_weights_only=False, mode='auto', period=1)

history=model.fit_generator(train_generator, samples_per_epoch= \
            len(train_samples), validation_data=validation_generator, \
            nb_val_samples=len(validation_samples), epochs=7, callbacks=[checkpoint])

'''
history=model.fit_generator(train_generator, steps_per_epoch=math.ceil(len(train_samples)/32), epochs=3,\
                    verbose=1, callbacks=[checkpoint], validation_data=validation_generator,\
                    validation_steps=math.ceil(len(validation_samples)/32), class_weight=None,\
                    max_queue_size=10, workers=1, use_multiprocessing=True, initial_epoch=0)



'''

#Evaluate the model with test_samples
print('Test Loss:', model.evaluate_generator(test_generator, 128))
#save the model and print the summary
model.save('model.h5') 
print("Model saved")
print(model.summary())
```

```python
import matplotlib.pyplot as plt
# list all data in history
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show();
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show();
```
